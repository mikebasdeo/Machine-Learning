# Simple Linear Regression Notes

Given a table with two columns, experience and salary, find the line of best fit between them.


## \begin{align*} 
y &= b_0 + b_1 \cdot x_1 \\
\text{salary} &= b_0 + b_1 \cdot \text{experience}
\end{align*}

### $y:$
Dependent varialble (Experience)
### $x_1:$
Independent variable (Salary)
### $b_0:$
The constant, aka the point where the line of best fit crosses the $y$-axis
### $b_1:$
Coefficient or proportion of change. The slope of the line.



### Ordinary Least Squares Method.
The minimum sum of squares.
